# Creating a logictic regresssion to predict absenteeism

## Import the relevant libraries

```python
import pandas as pd
import numpy as np
```

## Load the preprocessed dataset

```python
!pwd
```

```python
data_preprocessed = pd.read_csv('../Preprocessing/Absenteeism_preprocessed.csv')
```

## Create the targets

```python
data_preprocessed['Absenteeism Time in Hours'].median()
```

```python
targets = np.where(data_preprocessed['Absenteeism Time in Hours']>data_preprocessed['Absenteeism Time in Hours'].median(),1,0)
targets
```

```python
data_preprocessed['Excessive_absenteeism'] = targets
data_preprocessed.head()
```

```python
# note that a balance of 60 - 40 split is good !!!
```

```python
data_with_targets = data_preprocessed.drop('Absenteeism Time in Hours',axis = 1)
data_with_targets.head()
```

## Select the input for the regression

```python
data_with_targets.shape
```

```python
data_with_targets.iloc[:,:14]
```

```python
# Another and better way
unscaled_inputs = data_with_targets.iloc[:,:-1]
```

## Standardize the inputs
Let us choose the columns to scale.

```python
unscaled_inputs.columns.values
```

```python
columns_to_scale = ['Month value','Day of the week', 'Transportation Expense', 'Month value',
       'Age', 'Daily Work Load Average', 'Body Mass Index','Children', 'Pets', 'Education', 'Distance to Work']
```

```python
from sklearn.preprocessing import StandardScaler
absenteeism_scaler = StandardScaler()
```

```python
absenteeism_scaler.fit(unscaled_inputs[columns_to_scale])
```

```python
scaled_inputs = absenteeism_scaler.transform(unscaled_inputs[columns_to_scale])
```

```python
Reasons = unscaled_inputs[['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4']].to_numpy()
```

```python
scaled_inputs = np.concatenate((Reasons,scaled_inputs),axis=1)
```

```python
scaled_inputs
```

```python
scaled_inputs.shape
```

## Split the data into train and test and split the data

### Import the relevant modules

```python
from sklearn.model_selection import train_test_split
```

### Split

```python
train_test_split(scaled_inputs,targets)
```

```python
x_train, x_test, y_train, y_test = train_test_split(scaled_inputs,targets,train_size=0.8,random_state=20)
```

```python
print(x_train.shape,y_train.shape)
```

```python
print(x_test.shape,y_test.shape)
```

# Logistic regression with sklearn

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

### Training the model

```python
reg = LogisticRegression()
```

```python
reg.fit(x_train,y_train)
```

```python
reg.score(x_train,y_train)
```

```python
model_output = reg.predict(x_train)
model_output
```

```python
y_train
```

```python
model_output == y_train
```

```python
# Number of true entries
np.sum(model_output == y_train)
```

```python
model_output.shape[0]
```

```python
np.sum(model_output == y_train)/model_output.shape[0]
```

## Finding the intercept and coefficients

```python
reg.intercept_
```

```python
reg.coef_
```

```python
unscaled_inputs.columns.values
```

```python
feature_names = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4', 'Month value',
       'Day of the week', 'Transportation Expense', 'Body Mass Index','Children', 'Pets']
```

```python
summary_table = pd.DataFrame(columns=['Feature name'], data = columns_to_scale)
summary_table['Coefficient'] = np.transpose(reg.coef_)
summary_table
```

```python
summary_table.index = summary_table.index + 1
summary_table.loc[0] = ['intercept',reg.intercept_[0]]
summary_table     = summary_table.sort_index()
summary_table
```

# Interpreting the coefficients

```python
summary_table['Odd_ratio'] = np.exp(summary_table.Coefficient)
```

```python
summary_table
```

```python
summary_table.sort_values('Odd_ratio',ascending=False)
```

Note that coefficient around zero & odd ratio around one means that the feature is not particularly important
In that regard, `Daily Work Load Average`, `Distance to Work`, as well as `Day of the week`, `Education`, `Age 	`, seems to be the features with the most insignificant coefficients.
